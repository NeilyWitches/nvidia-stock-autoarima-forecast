```r
nvidia <- read.csv('C:/Users/drumm/Documents/practical time series analysis/nvidiastockprice5yrs.csv')
prices1 <- nvidia$Close.Last
prices2 <- rev(prices1)
prices3 <- as.character(prices2)
prices4 <- gsub(' \\$', '', prices3)
prices5 <- as.numeric(prices4)
prices6 <- ts(prices5)
```

```r
library(forecast)
```

```r
auto.arima(prices6)
```

```r
model<- auto.arima(prices6)
jpeg('autoarimaconfidence.jpg')
plot(forecast(model,h=252))
dev.off()
```

```r
library(astsa)
```

```r

a<-sarima.for(log(prices6),252,2,1,2)

jpeg('forecast.jpg')
plot.ts(c(prices6,exp(a$pred)), main='Daily stock prices + Forecast', ylab='', col='blue', lwd=3)
dev.off()
```
